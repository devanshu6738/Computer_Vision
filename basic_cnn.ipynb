```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
base_dir="/content/dogsvscats/train"
data_datagen=ImageDataGenerator(rescale=1./255,validation_split=0.2)
```

```python
train_gen=data_datagen.flow_from_directory(
    base_dir,
    target_size=(150,150),
    batch_size=50,
    class_mode='binary',
    subset='training'
)
valdation_gen=data_datagen.flow_from_directory(
    base_dir,
    target_size=(150,150),
    batch_size=50,
    class_mode='binary',
    subset='validation',
    shuffle=False
)
```

```python
train_gen.class_indices
```

```python
import keras
from keras import layers
model=keras.Sequential([
    # layer1
     layers.Conv2D(32,(3,3),activation='relu',input_shape=(150,150,3)),
     layers.MaxPooling2D(2,2),
    # layer2
     layers.Conv2D(64,(3,3),activation='relu',input_shape=(150,150,3)),
     layers.MaxPooling2D(2,2),
    # layer3
     layers.Conv2D(128,(3,3),activation='relu',input_shape=(150,150,3)),
     layers.MaxPooling2D(2,2),
    # layer4
     layers.Conv2D(128,(3,3),activation='relu',input_shape=(150,150,3)),
     layers.MaxPooling2D(2,2),

     layers.Flatten(),
     layers.Dense(512,activation='relu'),
     layers.Dense(1,activation='sigmoid')

])
model.summary()
```

```python
 model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(train_gen,epochs=10,validation_data=valdation_gen)
```

```python
test_datagen=ImageDataGenerator(rescale=1./255)
test_dir="/content/dogsvscats/test"
test_gen=test_datagen.flow_from_directory(
    test_dir,
    target_size=(150,150),
    batch_size=50,
    class_mode='binary',
    shuffle=False
)
```

```python
test_loss,test_accuracy=model.evaluate(test_gen,steps=len(test_gen))

print(test_loss)
print(test_accuracy)
```

```python
probabilities=model.predict(test_gen)
probabilities
```

```python
predicted_class=(probabilities>0.5).astype(int).squeeze()
predicted_class
```
