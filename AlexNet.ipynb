```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Activation, Dropout, BatchNormalization
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
base_dir="/content/cifar10/cifar10/train"
data_datagen=ImageDataGenerator(rescale=1./255,validation_split=0.2)
```

```python
train_gen=data_datagen.flow_from_directory(
    base_dir,
    target_size=(32,32),
    batch_size=128,
    class_mode='categorical',
    subset='training'
)
valdation_gen=data_datagen.flow_from_directory(
    base_dir,
    target_size=(32,32),
    batch_size=128,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)
```

```python
import keras
from keras import layers
model =keras.Sequential()

# Layer 1
model.add(Conv2D(96, kernel_size=(3,3), strides=(1,1), input_shape=(32,32,3), padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))
model.add(BatchNormalization())

# Layer 2
model.add(Conv2D(256, kernel_size=(3,3), strides=(1,1), padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))
model.add(BatchNormalization())

# Layer 3
model.add(Conv2D(384, kernel_size=(3,3), strides=(1,1), padding='same'))
model.add(Activation('relu'))

# Layer 4
model.add(Conv2D(384, kernel_size=(3,3), strides=(1,1), padding='same'))
model.add(Activation('relu'))

# Layer 5
model.add(Conv2D(256, kernel_size=(3,3), strides=(1,1), padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))

# Flatten
model.add(Flatten())

# Fully Connected Layer 1
model.add(Dense(1024))
model.add(Activation('relu'))
model.add(Dropout(0.5))

# Fully Connected Layer 2
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))

# Output Layer
model.add(Dense(10))
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
history=model.fit(train_gen,epochs=15,validation_data=valdation_gen,verbose=1,batch_size=128)
```

```python
test_dir="/content/cifar10/cifar10/test"
test_gen = data_datagen.flow_from_directory(
    test_dir,
    target_size=(32, 32),
    batch_size=128,
    class_mode='categorical',
    shuffle=False
)
test_loss, test_acc = model.evaluate(test_gen)

print("Test Accuracy:", test_acc)
print("Test Loss:", test_loss)
```
