Corner Detection (Harris Corner detection)

```python
import cv2 as cv
import numpy as np
```

```python
cap=cv.VideoCapture(0)

while True :
    res,frame=cap.read()
    if not res:
        print("camera error")
        break
    vd=cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    vd=np.float32(vd)

    des=cv.cornerHarris(vd,17,21,0.01)
    
    frame[des > 0.01 * des.max()] = [0, 0, 255]
    image_rgb = cv.cvtColor(frame, cv.COLOR_BGR2RGB)

    cv.imshow('Live Video',image_rgb)

    key=cv.waitKey(1) & 0xFF
    if key==ord('s'):
        break

cap.release()
cv.destroyAllWindows()
```

FAST Algorithm for corner detection

```python
cap=cv.VideoCapture(0)

while True :
    res,frame=cap.read()
    if not res:
        print("camera error")
        break
    vd=cv.cvtColor(frame,cv.COLOR_BGR2GRAY)

    fast = cv.FastFeatureDetector_create(threshold=50, nonmaxSuppression=True)
    kp = fast.detect(vd, None)

    frame_with_keypoints = cv.drawKeypoints(frame, kp, None, color=(0, 255, 0),flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    cv.imshow('Live Video',frame_with_keypoints)

    key=cv.waitKey(1) & 0xFF
    if key==ord('s'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
orb = cv.ORB_create(nfeatures=500,      # Max features to retain
                    scaleFactor=1.2,     # Pyramid decimation ratio
                    nlevels=8,           # Pyramid levels
                    edgeThreshold=31,    # Border size
                    firstLevel=0,        # Starting level
                    WTA_K=2,             # Pixels for BRIEF (2 or 3)
                    scoreType=cv.ORB_HARRIS_SCORE,  # ORB_HARRIS_SCORE or ORB_FAST_SCORE
                    patchSize=31,        # Patch size for descriptor
                    fastThreshold=20)    # FAST threshold
cap=cv.VideoCapture(0)
while True:
    res,frame=cap.read()

    if not res:
        print("camera error")
        break

    gray=cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    keypoints, descriptors = orb.detectAndCompute(gray, None)
    frame_with_kp = cv.drawKeypoints(frame, keypoints, None, color=(0, 255, 0),flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    frame_with_kp = cv.drawKeypoints(frame, keypoints, None, color=(0, 255, 0),flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

    cv.imshow('Live Video',frame_with_kp)

    if cv.waitKey(1) & 0xFF == ord('s'):
        break

cap.release()
cv.destroyAllWindows()
```
