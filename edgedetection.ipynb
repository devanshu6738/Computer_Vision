```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv.imread("images.png",cv.IMREAD_GRAYSCALE)
row,col=img.shape
```

```python
sobel_vertical=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
k=sobel_vertical.shape[0]
k
```

```python
output_matrix=np.zeros((row-k+1,col-k+1))
```

```python
for i in range(1,row-1):
    for j in range(1,col-1):
        region=img[i-1:i+2,j-1:j+2]
        output_matrix[i-1,j-1]=np.sum(sobel_vertical*region)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Original image")
plt.imshow(img,cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Sobel COnv Output")
plt.imshow(output_matrix,cmap='gray')
plt.axis('off')
```
