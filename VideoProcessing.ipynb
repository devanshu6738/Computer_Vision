```python
import cv2 as cv

cap=cv.VideoCapture(0)

while True:
    ret,frame=cap.read()
    
    if not ret:
        print("camera error")
        break
    cv.imshow('Video',frame)

    if cv.waitKey(1)==ord('s'):
        break

cap.release()
cv.destroyAllWindows()
```

task 1 - run the webcam - live video
task 2 - print frame shape
task 3 - convert webcam to grayscale
task 4 - flip the webcam horizontally like selfie camera

```python
# task1 and task2
cap=cv.VideoCapture(0)

while True:
    res,frame=cap.read()

    if not ret:
        print("camera error")
        break
    cv.imshow('Live Video',frame)
    print(frame.shape)
    if cv.waitKey(1)==ord('s'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
# task 3 
cap=cv.VideoCapture(0)

while True:
    res,frame=cap.read()

    if not ret:
        print("camera error")
        break
    gray=cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    cv.imshow('Live Video',gray)
    
    if cv.waitKey(1)==ord('s'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
# task 4

cap=cv.VideoCapture(0)

while True:
    res,frame=cap.read()
    frame=cv.flip(frame,1)

    if not res:
        print("camera error")
        break
    cv.imshow('Live Video',frame)

    if cv.waitKey(1)==ord('s'):
        break

cap.release()
cv.destroyAllWindows()
```
