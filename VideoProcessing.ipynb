```python
import cv2 as cv

cap=cv.VideoCapture(0)

while True:
    ret,frame=cap.read()
    
    if not ret:
        print("camera error")
        break
    cv.imshow('Video',frame)

    if cv.waitKey(1)==ord('s'):
        break

cap.release()
cv.destroyAllWindows()
```

task 1 - run the webcam - live video
task 2 - print frame shape
task 3 - convert webcam to grayscale

```python
# task1 and task2
cap=cv.VideoCapture(0)

while True:
    res,frame=cap.read()

    if not ret:
        print("camera error")
        break
    cv.imshow('Live Video',frame)
    print(frame.shape)
    if cv.waitKey(1)==ord('s'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
# task 3 
cap=cv.VideoCapture(0)

while True:
    res,frame=cap.read()

    if not ret:
        print("camera error")
        break
    gray=cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    cv.imshow('Live Video',gray)
    
    if cv.waitKey(1)==ord('s'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
# task 4

cap=cv.VideoCapture(0)

while True:
    res,frame=cap.read()
    frame=cv.flip(frame,1)

    if not res:
        print("camera error")
        break
    cv.imshow('Live Video',frame)

    if cv.waitKey(1)==ord('s'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
# get fps

import cv2 as cv

cap=cv.VideoCapture(0)

while True:
    res,frame=cap.read()
    if not res:
        print("camera error")
        break
    cv.imshow('Live Video',frame)
    fps=cap.get(cv.CAP_PROP_FPS)
    print(fps)

    if cv.waitKey(10000)==ord('s'):
        break

cap.release()
cv.destroyAllWindows()

    
```

```python
# capture a image and save into the file

import cv2 as cv

cap=cv.VideoCapture(0)
count=0
while(True):
    res,frame=cap.read()
    frame=cv.flip(frame,1)

    if not res:
        print("camera error")
        break
    cv.imshow('Live Video',frame)

    key=cv.waitKey(0) & 0xFF

    if key ==ord('c'):
        capture=frame
        filename=f"capture_{count}.jpg"
        cv.imwrite(filename,capture)
        count=count+1

    if key==ord('s'):
        break

cap.release()
cv.destroyAllWindows()
```
