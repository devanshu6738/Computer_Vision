```python
model=YOLO("yolov8n.pt")
```

```python
img=cv.imread("apple.jpg")
result=model(img)
```

```python
cap=cv.VideoCapture(0)

while True:
    res,frame=cap.read()
    if not res:
        print("camera error")
        break
    frame=cv.flip(frame,1)
    result=model(frame)
    annotated=result[0].plot()
    cv.imshow('Live Video',annotated)
    if cv.waitKey(1)==ord('s'):
        break

cap.release()
cv.destroyAllWindows()
    
```
