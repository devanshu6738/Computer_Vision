```python
import numpy as np
```

```python
image=np.array([
    [1,2,3,0,1],
    [0,1,2,3,2],
    [3,0,1,2,1],
    [2,3,1,0,0],
    [1,2,0,1,2]
])
kernel=np.array([
    [0,1,0],
    [1,-4,1],
    [0,1,0]
])
def conv(image,kernel):
    image_h,image_w=image.shape
    kernel_h,kernel_w=kernel.shape
    output_h=image_h-kernel_h+1
    output_w=image_w-kernel_w+1
    output_matrix=np.zeros((output_h,output_w))
    
    for i in range(output_h):
        for j in range(output_w):
            region=image[i:i+kernel_h,j:j+kernel_w]
            output_matrix[i,j]=np.sum(region*kernel)
            
    return output_matrix


output=conv(image,kernel)
        
output
```

```python
# padding

def padding(image,padd):
    output_h,output_w=image.shape
    output=np.zeros((output_h+padd*2,output_w+padd*2))
    
    for i in range(output_h):
        for j in range(output_w):
            output[i+padd,j+padd]=image[i,j]
            
    return output

output1=padding(image,1)

print(output1)

# numpy.pad

np.pad(image,1,mode='constant',constant_values=0)
```
